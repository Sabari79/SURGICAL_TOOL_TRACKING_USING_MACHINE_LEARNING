```python
import os
import requests
from duckduckgo_search import DDGS
from tqdm import tqdm
from PIL import Image
from io import BytesIO

# List of surgical tools
tools = ["Scalpel", "Straight Dissection Clamp", "Straight Mayo Scissor", "Curved Mayo Scissor"]
num_images = 20  # Number of images per tool

# Create a directory for images
save_dir = "surgical_tools"
os.makedirs(save_dir, exist_ok=True)

def scrape_images(tool_name, num_images):
    print(f"\nüîç Searching images for: {tool_name}...")
    
    # Create a subfolder for each tool
    tool_folder = os.path.join(save_dir, tool_name.replace(" ", "_"))
    os.makedirs(tool_folder, exist_ok=True)

    try:
        with DDGS() as ddgs:
            results = list(ddgs.images(tool_name, max_results=num_images))
        
        image_urls = [result["image"] for result in results if "image" in result]
    except Exception as e:
        print(f"‚ùå Failed to retrieve images: {e}")
        return  

    print(f"üîπ Found {len(image_urls)} images")
    
    for i, url in enumerate(tqdm(image_urls, desc=f"Downloading {tool_name} images")):
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()
            image = Image.open(BytesIO(response.content))
            
            # Save the image
            image_path = os.path.join(tool_folder, f"{tool_name.replace(' ', '_')}_{i+1}.jpg")
            image.save(image_path, "JPEG")
        except Exception as e:
            print(f"‚ùå Failed to download {url}: {e}")

    print(f"‚úÖ Downloaded {len(os.listdir(tool_folder))} images for {tool_name}.")

# Run the scraping process
for tool in tools:
    scrape_images(tool, num_images)

print("\nüéâ Image Scraping from DuckDuckGo Completed!")
```

```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from PIL import Image
from io import BytesIO

# Configuration
TOOLS = ["Scalpel", "Straight Dissection Clamp", "Straight Mayo Scissor", "Curved Mayo Scissor"]
IMAGE_LIMIT = 500  # Max images per tool
TARGET_RESOLUTION = (150, 150)  # Minimum resolution allowed

# Selenium Setup
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Function to fetch image URLs from Google
def fetch_google_images(query):
    print(f"üîç Searching {query} on Google Images...")
    
    search_url = f"https://www.google.com/search?tbm=isch&q={query.replace(' ', '+')}"
    driver.get(search_url)
    time.sleep(2)

    # Scroll down to load more images
    for _ in range(5):  
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
        time.sleep(2)

    # Extract image URLs
    image_urls = set()
    images = driver.find_elements(By.CSS_SELECTOR, "img")
    
    for img in images:
        try:
            src = img.get_attribute("src") or img.get_attribute("data-src")
            if src and "http" in src:
                image_urls.add(src)
        except:
            continue
    
    return list(image_urls)

# Function to download images with resolution check
def download_images(image_urls, folder_name):
    os.makedirs(folder_name, exist_ok=True)
    count = 0

    for url in image_urls:
        try:
            response = requests.get(url, timeout=5)
            image = Image.open(BytesIO(response.content))
            
            # Skip low-resolution images
            if image.size[0] < TARGET_RESOLUTION[0] or image.size[1] < TARGET_RESOLUTION[1]:
                print(f"‚ùå Skipped (low resolution {image.size}): {url}")
                continue  
            
            image_path = os.path.join(folder_name, f"{count}.jpg")
            image.save(image_path)
            count += 1
            print(f"‚úÖ Downloaded ({image.size}) -> {image_path}")
            
            if count >= IMAGE_LIMIT:
                break  
        except Exception as e:
            print(f"‚ö†Ô∏è Error downloading {url}: {e}")
    
    print(f"üéâ Downloaded {count} images for {folder_name}!")

# Main Execution
for tool in TOOLS:
    print(f"üîπ Processing {tool}...")
    images = fetch_google_images(tool)
    download_images(images, folder_name=f"dataset/{tool.replace(' ', '_')}")

driver.quit()
print("üöÄ Image Scraping Completed!")
```
